```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# Define constants
DELTA_T = 1  # Time increment
MAX_TIME = 100  # Maximum simulation time
GAMMA_GOAL = 1.0
GAMMA_COMBAT = 0.5
GAMMA_RECON = 0.2

# Ship types
SHIP_TYPES = {
    'Destroyer': {'speed': 30, 'firepower': 50, 'armor': 30, 'range': 100, 'recon': 80},
    'Cruiser': {'speed': 25, 'firepower': 70, 'armor': 50, 'range': 120, 'recon': 60},
    'Battleship': {'speed': 20, 'firepower': 100, 'armor': 100, 'range': 150, 'recon': 40},
    'AircraftCarrier': {'speed': 18, 'firepower': 30, 'armor': 60, 'range': 200, 'recon': 100}
}

# Helper functions
def distance(a, b):
    return math.hypot(a.x - b.x, a.y - b.y)

# Ship class
class Ship:
    def __init__(self, ship_id, ship_type, team, x, y):
        self.id = ship_id
        self.type = ship_type
        self.team = team
        self.x = x
        self.y = y
        self.lifeline = 100  # Health points
        self.weapon_level = SHIP_TYPES[ship_type]['firepower']
        self.attack_ability = SHIP_TYPES[ship_type]['firepower']
        self.attack_range = SHIP_TYPES[ship_type]['range']
        self.speed = SHIP_TYPES[ship_type]['speed']
        self.recon_ability = SHIP_TYPES[ship_type]['recon']
        self.direction = random.uniform(0, 2 * math.pi)  # Random initial direction

    def move(self):
        # Update position based on speed and direction
        self.x += self.speed * math.cos(self.direction) * DELTA_T
        self.y += self.speed * math.sin(self.direction) * DELTA_T

    def choose_action(self, enemy_ships):
        # Simple AI to choose an action
        # For demonstration, ships will randomly decide to move or attack if enemy in range
        action = {}
        # Check for enemies in range
        targets_in_range = [enemy for enemy in enemy_ships if distance(self, enemy) <= self.attack_range]
        if targets_in_range:
            # Attack the first enemy in range
            target = targets_in_range[0]
            action['type'] = 'attack'
            action['target'] = target
        else:
            # Move towards the closest enemy
            if enemy_ships:
                closest_enemy = min(enemy_ships, key=lambda enemy: distance(self, enemy))
                angle_to_enemy = math.atan2(closest_enemy.y - self.y, closest_enemy.x - self.x)
                self.direction = angle_to_enemy
            action['type'] = 'move'
        return action

    def attack(self, target):
        # Calculate damage
        damage = self.attack_ability - target.attack_ability * 0.5
        damage = max(damage, 0)
        target.lifeline -= damage
        if target.lifeline < 0:
            target.lifeline = 0

    def is_alive(self):
        return self.lifeline > 0

# Fleet class
class Fleet:
    def __init__(self, team_name):
        self.team_name = team_name
        self.ships = []

    def add_ship(self, ship):
        self.ships.append(ship)

    def get_alive_ships(self):
        return [ship for ship in self.ships if ship.is_alive()]

# Game class
class NavalBlockadeGame:
    def __init__(self):
        self.time = 0
        self.teams = {'Blockade': Fleet('Blockade'), 'Breakthrough': Fleet('Breakthrough')}
        self.history = defaultdict(list)

    def setup(self):
        # Initialize ships for both teams
        ship_id = 0
        # Blockade Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Blockade', random.uniform(0, 500), random.uniform(500, 1000))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Blockade', random.uniform(0, 500), random.uniform(500, 1000))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Blockade', random.uniform(0, 500), random.uniform(500, 1000))
        self.teams['Blockade'].add_ship(ship)
        ship_id += 1

        # Breakthrough Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Breakthrough', random.uniform(500, 1000), random.uniform(0, 500))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Breakthrough', random.uniform(500, 1000), random.uniform(0, 500))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Breakthrough', random.uniform(500, 1000), random.uniform(0, 500))
        self.teams['Breakthrough'].add_ship(ship)

    def run(self):
        self.setup()
        while self.time < MAX_TIME:
            print(f"Time: {self.time}")
            self.update()
            self.time += DELTA_T
            if self.check_end_conditions():
                break
        self.declare_winner()

    def update(self):
        # Each ship takes action
        for team_name, fleet in self.teams.items():
            alive_ships = fleet.get_alive_ships()
            enemy_team_name = 'Breakthrough' if team_name == 'Blockade' else 'Blockade'
            enemy_ships = self.teams[enemy_team_name].get_alive_ships()

            for ship in alive_ships:
                action = ship.choose_action(enemy_ships)
                if action['type'] == 'move':
                    ship.move()
                elif action['type'] == 'attack':
                    ship.attack(action['target'])

                # Record history for analysis
                self.history[ship.id].append((self.time, ship.x, ship.y, ship.lifeline))

    def check_end_conditions(self):
        # Check if any team has no ships left
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        if blockade_ships_alive == 0 or breakthrough_ships_alive == 0:
            return True
        return False

    def declare_winner(self):
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        if blockade_ships_alive > breakthrough_ships_alive:
            print("Blockade Team Wins!")
        elif breakthrough_ships_alive > blockade_ships_alive:
            print("Breakthrough Team Wins!")
        else:
            print("It's a Draw!")

    def plot_positions(self):
        # Plot the positions of ships over time
        for ship_id, records in self.history.items():
            xs = [record[1] for record in records]
            ys = [record[2] for record in records]
            plt.plot(xs, ys, label=f"Ship {ship_id}")
        plt.legend()
        plt.xlabel('X Position')
        plt.ylabel('Y Position')
        plt.title('Ship Movements Over Time')
        plt.show()

# Run the game simulation
if __name__ == "__main__":
    game = NavalBlockadeGame()
    game.run()
    game.plot_positions()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

class Ship:
    def __init__(self, x, y, team, ship_type):
        self.x = x
        self.y = y
        self.team = team
        self.ship_type = ship_type
        self.lifeline = 100
        self.weapon_level = np.random.randint(50, 100)
        self.attack_ability = np.random.randint(50, 100)
        self.attack_range = 50 if ship_type == 'destroyer' else 100
        self.speed = 5 if ship_type == 'destroyer' else 3

    def move(self, dx, dy):
        self.x += dx * self.speed
        self.y += dy * self.speed

    def attack(self, target):
        if self.in_range(target) and np.random.random() < self.attack_ability / 100:
            damage = self.weapon_level * (1 - target.attack_ability / 100)
            target.lifeline -= damage
            return damage
        return 0

    def in_range(self, target):
        distance = np.sqrt((self.x - target.x)**2 + (self.y - target.y)**2)
        return distance <= self.attack_range

class NavalBlockadeGame:
    def __init__(self, map_size=1000):
        self.map_size = map_size
        self.blockade_team = []
        self.breakthrough_team = []
        self.initialize_ships()

    def initialize_ships(self):
        # Blockade team
        for _ in range(3):
            self.blockade_team.append(Ship(np.random.randint(0, self.map_size),
                                           np.random.randint(self.map_size//2, self.map_size),
                                           'blockade', 'destroyer'))
        for _ in range(2):
            self.blockade_team.append(Ship(np.random.randint(0, self.map_size),
                                           np.random.randint(self.map_size//2, self.map_size),
                                           'blockade', 'battleship'))

        # Breakthrough team
        for _ in range(3):
            self.breakthrough_team.append(Ship(np.random.randint(0, self.map_size),
                                               np.random.randint(0, self.map_size//2),
                                               'breakthrough', 'destroyer'))
        for _ in range(2):
            self.breakthrough_team.append(Ship(np.random.randint(0, self.map_size),
                                               np.random.randint(0, self.map_size//2),
                                               'breakthrough', 'battleship'))

    def move_ships(self):
        for ship in self.blockade_team + self.breakthrough_team:
            if ship.team == 'blockade':
                ship.move(0, -1)  # Move down
            else:
                ship.move(0, 1)  # Move up
            ship.x = np.clip(ship.x, 0, self.map_size)
            ship.y = np.clip(ship.y, 0, self.map_size)

    def perform_attacks(self):
        for attacker in self.blockade_team + self.breakthrough_team:
            targets = self.breakthrough_team if attacker.team == 'blockade' else self.blockade_team
            for target in targets:
                attacker.attack(target)

    def remove_destroyed_ships(self):
        self.blockade_team = [ship for ship in self.blockade_team if ship.lifeline > 0]
        self.breakthrough_team = [ship for ship in self.breakthrough_team if ship.lifeline > 0]

    def check_game_end(self):
        if not self.breakthrough_team:
            return "Blockade Success"
        if not self.blockade_team:
            return "Blockade Failure"
        if any(ship.y >= self.map_size for ship in self.breakthrough_team):
            return "Blockade Failure"
        return None

    def step(self):
        self.move_ships()
        self.perform_attacks()
        self.remove_destroyed_ships()
        return self.check_game_end()

def init():
    ax.set_xlim(0, game.map_size)
    ax.set_ylim(0, game.map_size)
    return []

def animate(frame):
    ax.clear()
    result = game.step()
    
    for ship in game.blockade_team + game.breakthrough_team:
        color = 'red' if ship.team == 'blockade' else 'blue'
        marker = 'o' if ship.ship_type == 'destroyer' else 's'
        size = 100 * (ship.lifeline / 100)  # Size represents health
        ax.scatter(ship.x, ship.y, c=color, marker=marker, s=size, alpha=0.7)
        ax.text(ship.x, ship.y, f"{ship.lifeline:.0f}", ha='center', va='center', fontsize=8)

    ax.set_xlim(0, game.map_size)
    ax.set_ylim(0, game.map_size)
    ax.set_title(f"Naval Blockade Simulation - Step {frame}")
    
    if result:
        ax.text(game.map_size/2, game.map_size/2, result, ha='center', va='center', fontsize=20, bbox=dict(facecolor='white', alpha=0.5))
        anim.event_source.stop()
    
    return ax.collections + ax.texts

# Create game instance
game = NavalBlockadeGame()

# Set up the figure and animation
fig, ax = plt.subplots(figsize=(10, 10))
anim = FuncAnimation(fig, animate, init_func=init, frames=200, interval=200, blit=True, repeat=False)

plt.show()
```

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import matplotlib.animation as animation
import sys
import matplotlib
matplotlib.use('Agg')


# Import PyTorch for the DQN implementation
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Define constants
DELTA_T = 1  # Time increment
MAX_TIME = 100  # Maximum simulation time
GAMMA = 0.99  # Discount factor for RL
EPSILON = 0.1  # Exploration rate
LEARNING_RATE = 0.001
BATCH_SIZE = 32
MEMORY_SIZE = 1000

# Ship types
SHIP_TYPES = {
    'Destroyer': {'speed': 30, 'firepower': 50, 'armor': 30, 'range': 100, 'recon': 80},
    'Cruiser': {'speed': 25, 'firepower': 70, 'armor': 50, 'range': 120, 'recon': 60},
    'Battleship': {'speed': 20, 'firepower': 100, 'armor': 100, 'range': 150, 'recon': 40},
    'AircraftCarrier': {'speed': 18, 'firepower': 30, 'armor': 60, 'range': 200, 'recon': 100}
}

# Helper functions
def distance(a, b):
    return math.hypot(a.x - b.x, a.y - b.y)

def normalize_angle(angle):
    return angle % (2 * math.pi)

# DQN Network
class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.fc2 = nn.Linear(128, 128)
        self.out = nn.Linear(128, action_size)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.out(x)

# Replay Memory
class ReplayMemory:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
    
    def push(self, transition):
        self.memory.append(transition)
        if len(self.memory) > self.capacity:
            del self.memory[0]
    
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)

# Ship class
class Ship:
    def __init__(self, ship_id, ship_type, team, x, y):
        self.id = ship_id
        self.type = ship_type
        self.team = team
        self.x = x
        self.y = y
        self.lifeline = SHIP_TYPES[ship_type]['armor']
        self.weapon_level = SHIP_TYPES[ship_type]['firepower']
        self.attack_ability = SHIP_TYPES[ship_type]['firepower']
        self.attack_range = SHIP_TYPES[ship_type]['range']
        self.speed = SHIP_TYPES[ship_type]['speed']
        self.recon_ability = SHIP_TYPES[ship_type]['recon']
        self.direction = random.uniform(0, 2 * math.pi)  # Random initial direction
        self.memory = ReplayMemory(MEMORY_SIZE)
        self.policy_net = DQN(10, 5)  # State size and action size
        self.target_net = DQN(10, 5)
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=LEARNING_RATE)
        self.steps_done = 0
        self.last_state = None
        self.last_action = None
        self.color = 'blue' if self.team == 'Blockade' else 'red'
        self.alive = True
        self.recon_data = {}  # Known positions of enemy ships
    
    def get_state(self, enemy_ships):
        # State representation: [x, y, direction, lifeline, closest_enemy_x, closest_enemy_y, distance_to_enemy, enemy_lifeline, attack_range, speed]
        if enemy_ships:
            closest_enemy = min(enemy_ships, key=lambda enemy: distance(self, enemy))
            distance_to_enemy = distance(self, closest_enemy)
            state = np.array([
                self.x / 1000,
                self.y / 1000,
                self.direction / (2 * math.pi),
                self.lifeline / 100,
                closest_enemy.x / 1000,
                closest_enemy.y / 1000,
                distance_to_enemy / 1000,
                closest_enemy.lifeline / 100,
                self.attack_range / 200,
                self.speed / 30
            ])
        else:
            state = np.zeros(10)
        return state

    def select_action(self, state):
        sample = random.random()
        eps_threshold = EPSILON
        if sample > eps_threshold:
            with torch.no_grad():
                state_tensor = torch.from_numpy(np.array([state])).float()
                action_values = self.policy_net(state_tensor)
                return action_values.max(1)[1].item()
        else:
            return random.randrange(5)   
    
    def optimize_model(self):
        if len(self.memory) < BATCH_SIZE:
            return
        transitions = self.memory.sample(BATCH_SIZE)
        batch = list(zip(*transitions))
        
        state_batch = torch.tensor(batch[0], dtype=torch.float32)
        action_batch = torch.tensor(batch[1], dtype=torch.int64).unsqueeze(1)
        reward_batch = torch.tensor(batch[2], dtype=torch.float32)
        next_state_batch = torch.tensor(batch[3], dtype=torch.float32)
        done_batch = torch.tensor(batch[4], dtype=torch.float32)
        
        state_action_values = self.policy_net(state_batch).gather(1, action_batch)
        
        next_state_values = self.target_net(next_state_batch).max(1)[0].detach()
        expected_state_action_values = (next_state_values * GAMMA * (1 - done_batch)) + reward_batch
        
        loss = F.mse_loss(state_action_values.squeeze(), expected_state_action_values)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
    
    def move(self):
        # Update position based on speed and direction
        new_x = self.x + self.speed * math.cos(self.direction) * DELTA_T
        new_y = self.y + self.speed * math.sin(self.direction) * DELTA_T
        
        # Keep ships within bounds
        self.x = max(0, min(new_x, 1000))
        self.y = max(0, min(new_y, 1000))

        # Randomly change direction occasionally
        if random.random() < 0.1:  # 10% chance to change direction
            self.direction = random.uniform(0, 2 * math.pi)
    
    def perform_action(self, action, enemy_ships, allies):
        # Actions: 0 - Move Forward, 1 - Turn Left, 2 - Turn Right, 3 - Attack, 4 - Recon
        if action == 0:
            self.move()
        elif action == 1:
            self.direction = normalize_angle(self.direction - math.pi / 8)
            self.move()
        elif action == 2:
            self.direction = normalize_angle(self.direction + math.pi / 8)
            self.move()
        elif action == 3:
            self.attack(enemy_ships)
        elif action == 4:
            self.recon(enemy_ships, allies)
    
    def attack(self, enemy_ships):
        targets_in_range = [enemy for enemy in enemy_ships if distance(self, enemy) <= self.attack_range and enemy.alive]
        if targets_in_range:
            target = min(targets_in_range, key=lambda enemy: distance(self, enemy))
            hit_probability = self.attack_ability / 100
            if random.random() < hit_probability:
                damage = self.weapon_level
                target.lifeline -= damage
                if target.lifeline <= 0:
                    target.lifeline = 0
                    target.alive = False
            else:
                pass  # Missed the attack
    
    def recon(self, enemy_ships, allies):
        # Update own recon data
        for enemy in enemy_ships:
            if distance(self, enemy) <= self.recon_ability:
                self.recon_data[enemy.id] = (enemy.x, enemy.y)
        # Share recon data with allies
        for ally in allies:
            if ally != self:
                ally.recon_data.update(self.recon_data)
    
    def update(self, enemy_ships, allies):
        # Get current state
        state = self.get_state(enemy_ships)
        # Select and perform action
        action = self.select_action(state)
        self.perform_action(action, enemy_ships, allies)
        # Get reward
        reward = self.calculate_reward()
        # Get next state
        next_state = self.get_state(enemy_ships)
        # Check if done
        done = not self.alive
        # Store transition
        self.memory.push((state, action, reward, next_state, done))
        # Optimize model
        self.optimize_model()
        # Update last state and action
        self.last_state = next_state
        self.last_action = action
        # Update steps
        self.steps_done += 1
        # Periodically update target network
        if self.steps_done % 10 == 0:
            self.target_net.load_state_dict(self.policy_net.state_dict())
    
    def calculate_reward(self):
        # Reward function
        if not self.alive:
            return -1.0
        else:
            return 0.1  # Small reward for staying alive

    def is_alive(self):
        return self.alive

# Fleet class
class Fleet:
    def __init__(self, team_name):
        self.team_name = team_name
        self.ships = []
    
    def add_ship(self, ship):
        self.ships.append(ship)
    
    def get_alive_ships(self):
        return [ship for ship in self.ships if ship.is_alive()]
    
    def update(self, enemy_fleet):
        alive_ships = self.get_alive_ships()
        enemy_ships = enemy_fleet.get_alive_ships()
        for ship in alive_ships:
            ship.update(enemy_ships, alive_ships)

class NavalBlockadeGame:
    def __init__(self):
        self.time = 0
        self.teams = {'Blockade': Fleet('Blockade'), 'Breakthrough': Fleet('Breakthrough')}
        self.history = defaultdict(list)
        self.fig, self.ax = plt.subplots(figsize=(10, 10))
        self.user_control = False

    def setup(self):
        ship_id = 0
        # Blockade Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Blockade', random.uniform(200, 800), random.uniform(600, 800))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Blockade', random.uniform(200, 800), random.uniform(600, 800))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Blockade', random.uniform(200, 800), random.uniform(600, 800))
        self.teams['Blockade'].add_ship(ship)
        ship_id += 1

        # Breakthrough Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Breakthrough', random.uniform(200, 800), random.uniform(200, 400))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Breakthrough', random.uniform(200, 800), random.uniform(200, 400))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Breakthrough', random.uniform(200, 800), random.uniform(200, 400))
        self.teams['Breakthrough'].add_ship(ship)

    def run(self):
        self.setup()
        if self.user_control:
            self.user_strategy_input()
        
        ani = animation.FuncAnimation(self.fig, self.update_frame, frames=MAX_TIME, 
                                      interval=100, blit=False, repeat=False)
        
        ani.save('naval_blockade_animation.gif', writer='pillow', fps=10)
        print("Animation saved as 'naval_blockade_animation.gif'")
        
        plt.close(self.fig)
        
        self.declare_winner()

    def update_frame(self, frame):
        self.ax.clear()
        self.ax.set_xlim(0, 1000)
        self.ax.set_ylim(0, 1000)
        self.ax.set_title(f'Naval Blockade Simulation - Time: {self.time:.1f}')
    
        all_ships = self.teams['Blockade'].ships + self.teams['Breakthrough'].ships
    
        for team_name, fleet in self.teams.items():
            for ship in fleet.ships:
                if ship.is_alive():
                   ship.update(all_ships, fleet.get_alive_ships())
            
                color = ship.color if ship.is_alive() else 'gray'
                self.ax.plot(ship.x, ship.y, marker='o', color=color, markersize=10)
                self.ax.text(ship.x, ship.y, f"{ship.type[0]}:{ship.lifeline:.1f}", color=color)
                self.history[ship.id].append((self.time, ship.x, ship.y, ship.lifeline))
            
            # Plot ship's path
                path = self.history[ship.id]
                x = [p[1] for p in path]
                y = [p[2] for p in path]
                self.ax.plot(x, y, color=color, alpha=0.5)
    
    # Correct legend
        self.ax.plot([], [], color='blue', label='Blockade')
        self.ax.plot([], [], color='red', label='Breakthrough')
        self.ax.plot([], [], color='gray', label='Destroyed')
        self.ax.legend()

        self.time += DELTA_T
    
        if self.check_end_conditions():
           plt.close(self.fig)
        

    def check_end_conditions(self):
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        return blockade_ships_alive == 0 or breakthrough_ships_alive == 0

    def declare_winner(self):
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        if blockade_ships_alive > breakthrough_ships_alive:
            print("Blockade Team Wins!")
        elif breakthrough_ships_alive > blockade_ships_alive:
            print("Breakthrough Team Wins!")
        else:
            print("It's a Draw!")

    def user_strategy_input(self):
        print("User Control Enabled for Blockade Team.")
        for ship in self.teams['Blockade'].ships:
            print(f"Set direction for Ship {ship.id} ({ship.type}):")
            angle = float(input("Enter angle in degrees (0-360): "))
            ship.direction = math.radians(angle)

# Run the game simulation
if __name__ == "__main__":
    game = NavalBlockadeGame()
    if len(sys.argv) > 1 and sys.argv[1] == 'user':
        game.user_control = True
    game.run()

```

```python
import math
import random
import numpy as np
import sys
import pygame
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Initialize Pygame
pygame.init()

# Screen dimensions
SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 800

# Define constants
DELTA_T = 1  # Time increment
MAX_TIME = 100  # Maximum simulation time
GAMMA = 0.99  # Discount factor for RL
EPSILON = 0.1  # Exploration rate
LEARNING_RATE = 0.001
BATCH_SIZE = 32
MEMORY_SIZE = 1000

# Ship types
SHIP_TYPES = {
    'Destroyer': {'speed': 30, 'firepower': 50, 'armor': 30, 'range': 100, 'recon': 80, 'image': 'destroyer.jpeg'},
    'Cruiser': {'speed': 25, 'firepower': 70, 'armor': 50, 'range': 120, 'recon': 60, 'image': 'cruiser.jpeg'},
    'Battleship': {'speed': 20, 'firepower': 100, 'armor': 100, 'range': 150, 'recon': 40, 'image': 'battleship.jpeg'},
    'AircraftCarrier': {'speed': 18, 'firepower': 30, 'armor': 60, 'range': 200, 'recon': 100, 'image': 'carrier.jpeg'}
}

# Colors
BLUE = (0, 0, 255)
RED = (255, 0, 0)
GRAY = (128, 128, 128)
WHITE = (255, 255, 255)

# Helper functions
def distance(a, b):
    return math.hypot(a.x - b.x, a.y - b.y)

def normalize_angle(angle):
    return angle % (2 * math.pi)

# DQN Network
class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.fc2 = nn.Linear(128, 128)
        self.out = nn.Linear(128, action_size)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.out(x)

# Replay Memory
class ReplayMemory:
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
    
    def push(self, transition):
        self.memory.append(transition)
        if len(self.memory) > self.capacity:
            del self.memory[0]
    
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)

# Ship class
class Ship:
    def __init__(self, ship_id, ship_type, team, x, y):
        self.id = ship_id
        self.type = ship_type
        self.team = team
        self.x = x
        self.y = y
        self.lifeline = SHIP_TYPES[ship_type]['armor']
        self.weapon_level = SHIP_TYPES[ship_type]['firepower']
        self.attack_ability = SHIP_TYPES[ship_type]['firepower']
        self.attack_range = SHIP_TYPES[ship_type]['range']
        self.speed = SHIP_TYPES[ship_type]['speed']
        self.recon_ability = SHIP_TYPES[ship_type]['recon']
        self.direction = random.uniform(0, 2 * math.pi)  # Random initial direction
        self.memory = ReplayMemory(MEMORY_SIZE)
        self.policy_net = DQN(10, 5)  # State size and action size
        self.target_net = DQN(10, 5)
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=LEARNING_RATE)
        self.steps_done = 0
        self.last_state = None
        self.last_action = None
        self.alive = True
        self.recon_data = {}  # Known positions of enemy ships
        self.image = pygame.image.load(SHIP_TYPES[ship_type]['image']).convert_alpha()
        self.rect = self.image.get_rect(center=(self.x, self.y))
        self.color = BLUE if self.team == 'Blockade' else RED
    
    def get_state(self, enemy_ships):
        # State representation
        if enemy_ships:
            closest_enemy = min(enemy_ships, key=lambda enemy: distance(self, enemy))
            distance_to_enemy = distance(self, closest_enemy)
            state = np.array([
                self.x / SCREEN_WIDTH,
                self.y / SCREEN_HEIGHT,
                self.direction / (2 * math.pi),
                self.lifeline / 100,
                closest_enemy.x / SCREEN_WIDTH,
                closest_enemy.y / SCREEN_HEIGHT,
                distance_to_enemy / 1000,
                closest_enemy.lifeline / 100,
                self.attack_range / 200,
                self.speed / 30
            ])
        else:
            state = np.zeros(10)
        return state
    
    def select_action(self, state):
        sample = random.random()
        eps_threshold = EPSILON
        if sample > eps_threshold:
            with torch.no_grad():
                state_tensor = torch.from_numpy(np.array([state])).float()
                action_values = self.policy_net(state_tensor)
                return action_values.max(1)[1].item()
        else:
            return random.randrange(5)
    
    def optimize_model(self):
        if len(self.memory) < BATCH_SIZE:
            return
        transitions = self.memory.sample(BATCH_SIZE)
        batch = list(zip(*transitions))
        
        state_batch = torch.tensor(batch[0], dtype=torch.float32)
        action_batch = torch.tensor(batch[1], dtype=torch.int64).unsqueeze(1)
        reward_batch = torch.tensor(batch[2], dtype=torch.float32)
        next_state_batch = torch.tensor(batch[3], dtype=torch.float32)
        done_batch = torch.tensor(batch[4], dtype=torch.float32)
        
        state_action_values = self.policy_net(state_batch).gather(1, action_batch)
        
        next_state_values = self.target_net(next_state_batch).max(1)[0].detach()
        expected_state_action_values = (next_state_values * GAMMA * (1 - done_batch)) + reward_batch
        
        loss = F.mse_loss(state_action_values.squeeze(), expected_state_action_values)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
    
    def move(self):
        # Update position based on speed and direction
        new_x = self.x + self.speed * math.cos(self.direction) * DELTA_T
        new_y = self.y + self.speed * math.sin(self.direction) * DELTA_T
        
        # Keep ships within bounds
        self.x = max(0, min(new_x, SCREEN_WIDTH))
        self.y = max(0, min(new_y, SCREEN_HEIGHT))
        
        # Update rect position
        self.rect.center = (self.x, self.y)
    
    def perform_action(self, action, enemy_ships, allies):
        # Actions: 0 - Move Forward, 1 - Turn Left, 2 - Turn Right, 3 - Attack, 4 - Recon
        if action == 0:
            self.move()
        elif action == 1:
            self.direction = normalize_angle(self.direction - math.pi / 8)
            self.move()
        elif action == 2:
            self.direction = normalize_angle(self.direction + math.pi / 8)
            self.move()
        elif action == 3:
            self.attack(enemy_ships)
        elif action == 4:
            self.recon(enemy_ships, allies)
    
    def attack(self, enemy_ships):
        targets_in_range = [enemy for enemy in enemy_ships if distance(self, enemy) <= self.attack_range and enemy.alive]
        if targets_in_range:
            target = min(targets_in_range, key=lambda enemy: distance(self, enemy))
            hit_probability = self.attack_ability / 100
            if random.random() < hit_probability:
                damage = self.weapon_level
                target.lifeline -= damage
                if target.lifeline <= 0:
                    target.lifeline = 0
                    target.alive = False
            else:
                pass  # Missed the attack
    
    def recon(self, enemy_ships, allies):
        # Update own recon data
        for enemy in enemy_ships:
            if distance(self, enemy) <= self.recon_ability:
                self.recon_data[enemy.id] = (enemy.x, enemy.y)
        # Share recon data with allies
        for ally in allies:
            if ally != self:
                ally.recon_data.update(self.recon_data)
    
    def update(self, enemy_ships, allies):
        # Get current state
        state = self.get_state(enemy_ships)
        # Select and perform action
        action = self.select_action(state)
        self.perform_action(action, enemy_ships, allies)
        # Get reward
        reward = self.calculate_reward()
        # Get next state
        next_state = self.get_state(enemy_ships)
        # Check if done
        done = not self.alive
        # Store transition
        self.memory.push((state, action, reward, next_state, done))
        # Optimize model
        self.optimize_model()
        # Update last state and action
        self.last_state = next_state
        self.last_action = action
        # Update steps
        self.steps_done += 1
        # Periodically update target network
        if self.steps_done % 10 == 0:
            self.target_net.load_state_dict(self.policy_net.state_dict())
    
    def calculate_reward(self):
        # Reward function
        if not self.alive:
            return -1.0
        else:
            return 0.1  # Small reward for staying alive
    
    def is_alive(self):
        return self.alive

    def draw(self, screen):
        if self.is_alive():
            rotated_image = pygame.transform.rotate(self.image, -math.degrees(self.direction))
            new_rect = rotated_image.get_rect(center=self.rect.center)
            screen.blit(rotated_image, new_rect.topleft)
            # Draw lifeline
            lifeline_ratio = self.lifeline / SHIP_TYPES[self.type]['armor']
            pygame.draw.rect(screen, GRAY, (self.x - 20, self.y - 40, 40, 5))
            pygame.draw.rect(screen, self.color, (self.x - 20, self.y - 40, 40 * lifeline_ratio, 5))

# Fleet class
class Fleet:
    def __init__(self, team_name):
        self.team_name = team_name
        self.ships = []
    
    def add_ship(self, ship):
        self.ships.append(ship)
    
    def get_alive_ships(self):
        return [ship for ship in self.ships if ship.is_alive()]
    
    def update(self, enemy_fleet):
        alive_ships = self.get_alive_ships()
        enemy_ships = enemy_fleet.get_alive_ships()
        for ship in alive_ships:
            ship.update(enemy_ships, alive_ships)
    
    def draw(self, screen):
        for ship in self.ships:
            ship.draw(screen)

class NavalBlockadeGame:
    def __init__(self):
        self.time = 0
        self.teams = {'Blockade': Fleet('Blockade'), 'Breakthrough': Fleet('Breakthrough')}
        self.history = defaultdict(list)
        self.user_control = False
        self.screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
        pygame.display.set_caption('Naval Blockade Simulation')
        self.clock = pygame.time.Clock()
        self.font = pygame.font.SysFont('Arial', 16)
        self.running = True
        self.background = pygame.Surface(self.screen.get_size()).convert()
        self.background.fill((0, 105, 148))  # Ocean blue color

    def setup(self):
        ship_id = 0
        # Blockade Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Blockade', random.uniform(100, 900), random.uniform(600, 750))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Blockade', random.uniform(100, 900), random.uniform(600, 750))
            self.teams['Blockade'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Blockade', random.uniform(100, 900), random.uniform(600, 750))
        self.teams['Blockade'].add_ship(ship)
        ship_id += 1

        # Breakthrough Team Ships
        for _ in range(3):
            ship = Ship(ship_id, 'Destroyer', 'Breakthrough', random.uniform(100, 900), random.uniform(50, 200))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        for _ in range(2):
            ship = Ship(ship_id, 'Cruiser', 'Breakthrough', random.uniform(100, 900), random.uniform(50, 200))
            self.teams['Breakthrough'].add_ship(ship)
            ship_id += 1
        ship = Ship(ship_id, 'Battleship', 'Breakthrough', random.uniform(100, 900), random.uniform(50, 200))
        self.teams['Breakthrough'].add_ship(ship)

    def run(self):
        self.setup()
        if self.user_control:
            self.user_strategy_input()
        
        while self.running and self.time < MAX_TIME:
            self.handle_events()
            self.update()
            self.draw()
            pygame.display.flip()
            self.clock.tick(30)  # Limit to 30 FPS
            self.time += DELTA_T
            if self.check_end_conditions():
                self.running = False
        
        self.declare_winner()
        pygame.quit()
    
    def handle_events(self):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                self.running = False
    
    def update(self):
        # Update fleets
        self.teams['Blockade'].update(self.teams['Breakthrough'])
        self.teams['Breakthrough'].update(self.teams['Blockade'])
        # Record history
        for team_name, fleet in self.teams.items():
            for ship in fleet.ships:
                self.history[ship.id].append((self.time, ship.x, ship.y, ship.lifeline))
    
    def draw(self):
        self.screen.blit(self.background, (0, 0))
        # Draw fleets
        self.teams['Blockade'].draw(self.screen)
        self.teams['Breakthrough'].draw(self.screen)
        # Draw game info
        time_text = self.font.render(f"Time: {self.time}", True, WHITE)
        self.screen.blit(time_text, (10, 10))
    
    def check_end_conditions(self):
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        return blockade_ships_alive == 0 or breakthrough_ships_alive == 0

    def declare_winner(self):
        blockade_ships_alive = len(self.teams['Blockade'].get_alive_ships())
        breakthrough_ships_alive = len(self.teams['Breakthrough'].get_alive_ships())
        if blockade_ships_alive > breakthrough_ships_alive:
            winner_text = "Blockade Team Wins!"
        elif breakthrough_ships_alive > blockade_ships_alive:
            winner_text = "Breakthrough Team Wins!"
        else:
            winner_text = "It's a Draw!"
        # Display winner
        print(winner_text)
        # Show winner on screen
        winner_surface = self.font.render(winner_text, True, WHITE)
        self.screen.blit(winner_surface, (SCREEN_WIDTH // 2 - winner_surface.get_width() // 2, SCREEN_HEIGHT // 2))
        pygame.display.flip()
        pygame.time.delay(5000)
    
    def user_strategy_input(self):
        print("User Control Enabled for Blockade Team.")
        for ship in self.teams['Blockade'].ships:
            print(f"Set direction for Ship {ship.id} ({ship.type}):")
            angle = float(input("Enter angle in degrees (0-360): "))
            ship.direction = math.radians(angle)

# Run the game simulation
if __name__ == "__main__":
    game = NavalBlockadeGame()
    if len(sys.argv) > 1 and sys.argv[1] == 'user':
        game.user_control = True
    game.run()
```
